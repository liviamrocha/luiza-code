### **Tipos das Variáveis**

#### **Variáveis Numéricas**

Dizemos que uma variável é numérica quando armazena números inteiros ou
de ponto flutuante.

Os números inteiros são aqueles sem parte decimal, como 10, 0, 65, -8, 9898, -537.

Números de ponto flutuante são aqueles com parte decimal, como
1.0, 5.478, 10.478, -5000.4. Observe que 1.0, mesmo tendo zero na parte decimal, é um número de ponto flutuante. 

**OBS: Em Python, utilizamos o ponto, e não a vírgula, como separador entre a parte inteira e a fracionária de um número.**

**Exemplo:**

```
# Exemplos de variáveis numéricas

# VARIÁVEIS INTEIRAS
a = 10
b = 285
c = -98

# VARIÁVEIS DE PONTO FLUTUANTE
x = 10.2
y = -24.5
z = 2.0

```

#### **Variáveis do Tipo Lógico**

Muitas vezes, queremos armazenar um conteúdo simples: **verdadeiro** ou **falso** em uma variável. Nesse caso, utilizamos um tipo de variável chamado tipo lógico
ou booleano. Em Python, escrevemos "`True`" para verdadeiro e "`False`" para falso

**OBS: Observe que o T de "True" e o F de "False" são escritos com letras maiúsculas**

**Exemplo:**

```
#Execute o código colocando a variável "feio" na função print

lindo = True
feio = False
print(lindo)
```

Variáveis de tipo lógico serão muito utilizadas quando formos aprender sobre estruturas condicionais!

#### **Variáveis do Tipo String**

Variáveis do tipo string armazenam nomes e textos que podem ser considerados como cadeias de caracteres. Chamamos cadeia de caracteres uma sequência de símbolos como letras, números, sinais de pontuação etc. 

**Exemplo:** "`João e Maria comem pão.`" 

Nesse caso, João é uma sequência com as letras J, o, ã, o. Para simplificar o texto, utilizaremos o nome string para mencionar cadeias de caracteres. Podemos imaginar uma string como uma sequência de blocos, onde cada letra, número ou espaço em branco ocupa uma posição.

Uma string é reconhecida através das aspas (") que são colocadas no início e no fim da sequência de caracteres a fim de realizar sua demarcação.

**Exemplo:**

```
#Veja que nesse caso não há qualquer problema em utilizar espaço
frase1 = "João e Maria comem pão" 
frase2 = "Amor é fogo que arde sem se ver"
frase3 = "Silêncio"
```

Perceba que no código acima a string "João e Maria comem pão" ficou destacada na cor vermelha. Isso se chama "highlight", o Google Colab entende que o programador está criando uma string e facilita a leitura do conteúdo da variável alterando a cor do mesmo. Essa técnica também é aplicada para variáveis booleanas, comentários, funções etc. Obviamente, cada interpretador define uma ordem diferentes de cores.

**Concatenação de Strings**

O conteúdo de variáveis string podem ser juntados, ou melhor, concatenados.
Para concatenar duas strings, utilizamos o operador de adição (+). Assim, se concatenarmos as strings "AB" e "C" em uma operação `"AB" + "C"` temos como resultado a string `"ABC"`.

**Exemplo:**

```
ab = "AB"
c = "C"
resultado = ab + c
print(resultado)
```

A concatenação de strings pode ser realizada dentro da função print

```
print("AB" + "C")
```

**Formatação de Strings**

É uma generalização da concatenação de strings. Agora, podemos juntar uma string com uma variável de qualquer tipo! 

Isso ocorre porque a formatação de strings faz uma conversão automática do tipo da variável que será formatada para uma string.

Para formatar uma string, devemos preceder aquilo que está dentro das aspas com a letra **f**. Dentro das aspas, abrimos chaves na posição da string que desejamos que seja formatada.

Dentro das chaves, podemos inserir uma variável de qualquer tipo.

```
idade = 21
mensagem = f"olá, eu tenho {idade} anos"
print(mensagem)
```

Assim como na concatenação, a formatação também pode ser realizada dentro da função print.

```
print(f"olá, eu tenho {idade} anos")
```

### **Precedência de operações**

Os parênteses são utilizados em Python da mesma forma que em expressões
matemáticas, ou seja, para alterar a ordem de execução de uma operação. Para
relembrar a ordem de precedência das operações, temos as seguintes prioridades:

1. Exponenciação ou potenciação (**).
2. Multiplicação (*) e divisão (/, // e %).
3. Adição (+) e subtração (-).

**Exemplos:**

```
(10 - 8) * 3
```

Perceba que o resultado da operação acima é 6, pois os parênteses alteraram a ordem de execução da operação.

```
3 * 4 ** 2
```

O resultado da célula acima foi 48 pois a **exponenciação** possui precedência maior que a **multiplicação**.

```
10 / 2 * 5
```

Quando temos uma expressão com operadores com a mesma ordem de precedência a operação é feita da esquerda para a direita. Esse é o motivo pelo qual a célula resultou em 25.0.

Um programa é executado linha por linha pelo computador, executando as
operações descritas no programa uma após a outra. Quando trabalhamos com
variáveis, devemos nos lembrar de que o conteúdo de uma variável pode mudar
com o tempo. Isso porque a cada vez que alteramos o valor de uma variável, o
valor anterior é substituído pelo novo.

**Exemplo:**

```
a = 10      #Nessa linha a vale 10 e b ainda não foi declarado
b = a       #Nessa linha tanto b quanto a valem 10
a = 100     #Nessa linha b vale 10, mas a agora vale 100
print(b)
```

Se alterarmos a ordem de execução, **b** pode ter seu valor alterado na função **print**

**Exemplo:**

```
#Nesse caso o valor de a é modificado antes que a função print revelando o valor de b seja chamada

a = 10      
a = 100       
b = a     
print(b)
```

Uma das principais diferenças entre ler um texto e um programa é justamente
seguir as mudanças de valores de cada variável conforme o programa é executado.
Entender que o valor das variáveis pode mudar durante a execução do programa
não é tão natural, mas é fundamental para a programação de computadores. Um
programa não pode ser lido como um texto, mas cuidadosamente analisado linha
a linha. Ao escrever seus programas, verifique linha a linha os efeitos e mudanças causados no valor de cada variável, essa técnica é chamada de **rastreamento** e é fundamental para a leitura correta de um programa.

## **Entrada de dados**

Chamamos de entrada de dados o momento em que seu programa recebe dados ou valores por meio de um dispositivo de entrada de dados (como o teclado do computador) ou de um arquivo em disco. Até agora aprendemos a programar de uma forma que estávamos limitados a declaração de variáveis dentro do programa, agora que aprenderemos como funciona a entrada de dados, a interação entre o programa e o usuário será facilitada.

```
#Execute o exemplo para ver como funciona a função INPUT

x = input("Digite um número: ") 
print(x)
```

A função input é utilizada para solicitar dados do usuário. Ela recebe um parâmetro, que é a mensagem a ser exibida, e retorna o valor digitado pelo usuário.

**Exemplo:**

```
nome = input("Digite seu nome:")
print("Você digitou " + nome)
print("Olá, " + nome)
```

### **Conversão de entrada de dados**

A função input sempre retorna valores do tipo string, ou seja, não importa se digitamos apenas números, o resultado sempre é de tipo string. Isso pode ocasionar um problema, pois operações aritiméticas não podem ser realizadas entre strings. 

```
numero = input("Digite um número: ")
type(numero)
```

Para resolver esse pequeno problema, vamos aplicar a conversão de tipos utilizando a função **int** para converter o valor retornado em um número inteiro, e a função **float** para convertê-lo em número decimal ou de ponto flutuante.

**Exemplo:**

```
#Calculando o bônus por tempo de serviço

anos = int(input("Anos de serviço: "))
valor_por_ano = float(input("Valor por ano: "))
bonus = anos * valor_por_ano
print(f"Bônus de R$ {bonus}")
```

### **Realizando uma entrada de dois ou mais dados de uma única vez**

Em alguns exercícios durante o curso, você vai ser requisitado para realizar a entrada de dois ou mais dados numa única linha. Isso pode ser feito a partir do método **split**!

Um método é uma função que realiza alguma ação. Contundo, veremos isso mais para o final do curso. Podemos chamar o método **split** para um input da seguinte forma:

```
x, y = input("Digite dois dados: ").split()
print(f"x = {x}")
print(f"y = {y}")
```

Note que os dados devem sempre estar separados por um espaço. Nesse caso, primeiro foi escrito "10" (que foi atribuído a variável x), depois um espaço, e depois foi escrito "banana", que foi atribuída a variável y.

Além disso, é bom notar que os tipos tanto de x quanto de y continuam strings, pois não foi feita a conversão de tipos. Como geralmente queremos que "10" seja convertido para inteiro, devemos fazer isso manualmente:

```
x = int(x)
type(x), type(y)
```

### Reatribuição de variáveis

Há momentos que vamos sentir a necessidade de alterar valores de algumas variáveis, como por exemplo: substituir o resultado de uma soma, uma contagem, um nome... 

Sendo assim, podemos oferecer novos valores às variáveis que declaramos anteriormente apenas chamando-as novamente e atribuindo o seu novo valor.

```
x = 5 # valor inicial
print(x)
x = 7 # o valor anterior foi descartado e a variável recebeu um novo valor
print(x)
```

E assim como a atribuição de novos valores a uma variável, também podemos atribuir valores de uma variável a outra variável apenas informando que, por exemplo, a variável x recebe o valor armazenado na variável y.

```
a = 10
b = 3
print(f"O valor de a é:{a}")
print(f"O valor de b é: {b}")
b = a
print(f"O valor de a continua a ser: {a}")
print(f"E o novo valor de b é: {b}")
```

## **Saída de Dados**

Diversas vezes é necessário que os dados armazenados e manipulados no programa sejam exibidos para um usuário. Isso é feito por meio da função **print** que já utilizamos inúmeras vezes.

**Exemplos:**

```
nome = input("Insira seu nome: ")
diaria = float(input("Insira qual é sua remuneração diária (em R$): "))
dias_trabalhados = int(input("Insira quanto dias você trabalhou durante o mês: "))

print(f'Você se chama {nome} e recebe R$ {dias_trabalhados * diaria}')
```

Quando falamos de **saída de dados** os termos **imprima** e **escreva** são recorrentes.

Também é possível imprimirmos mais de um dado. **Por exemplo:**

```
nome = "Rebeca"
idade = 19
profissao = "Engenheira de Dados"

# Os dados são impressos na mesma linha
print(nome, idade, profissao) 
```

## **Dicas Extras**

Podemos simplificar algumas ações durante o código. A seguir estão algumas simplifações para as atribuições:

`x = x + 1` pode ser escrito como `x += 1` 

`x = x - 1` pode ser escrito como `x -= 1`

`x = x / 2` pode ser escrito como `x /= 2`

`x = x // 3` pode ser escrito como `x //= 3`

`x = x * 2` pode ser escrito como `x *= 2`

Ambos fazem a mesma coisa. A segunda opção tende a ser mais usada pela praticidade.
